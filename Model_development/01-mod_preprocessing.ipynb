# Plotting a random sample (raw)

```python
# Path to raw EEG folder
raw_folder = 'Raw_eeg_only'
raw_files = [f for f in os.listdir(raw_folder) if f.endswith('.raw')]

for file in raw_files:
    file_path = os.path.join(raw_folder, file)
    print(f"\nLoading: {file_path}")
    
    # Load raw data
    raw = mne.io.read_raw_egi(file_path, preload=True)
    
    # Only keep EEG channels
    raw.pick_types(eeg=True, stim=False)  # drop stim channels from view
    
    # Apply bandpass filter (1â€“45 Hz)
    raw.filter(1., 45., fir_design='firwin')

    # Drop E129 if present
    if 'E129' in raw.ch_names:
        raw.drop_channels(['E129'])

    # Set montage (HydroCel 129)
    montage = mne.channels.make_standard_montage('GSN-HydroCel-129')
    raw.set_montage(montage, on_missing='ignore')

    # Plot 30 EEG channels (just a 5-sec window to test)
    raw.plot(n_channels=30, duration=5.0, scalings='auto', block=True, title=f"{file}")
    
    break  # Just test with the first sample
```

# store event handler(epochs) in separate fif files to assess brain activity during different events

```python
import mne
import os

# Folder containing your .raw files
data_folder = 'Raw_eeg_only'
output_folder = 'Epochs_output'
os.makedirs(output_folder, exist_ok=True)

# Loop through each .raw file
for filename in os.listdir(data_folder):
    if filename.endswith('.raw'):
        filepath = os.path.join(data_folder, filename)
        print(f"\nProcessing: {filename}")

        # Load raw EEG
        raw = mne.io.read_raw_egi(filepath, preload=True)

        # Apply bandpass filter
        raw.filter(1., 45., fir_design='firwin')

        # Set montage (handling E129 separately to avoid errors)
        if 'E129' in raw.ch_names:
             raw.drop_channels(['E129'])

        montage = mne.channels.make_standard_montage('GSN-HydroCel-129')
        raw.set_montage(montage)


        # Extract events from annotations
        events, event_id = mne.events_from_annotations(raw)

        # Filter event IDs to only include cue-related triggers (or choose what you want)
        cue_event_ids = {k: v for k, v in event_id.items() if 'cue' in k}

        if not cue_event_ids:
            print(f"No 'cue' events found in {filename}. Skipping.")
            continue

        # Epoch the data (e.g., -0.2 to 0.8s around event)
        try:
            epochs = mne.Epochs(
                raw,
                events,
                event_id=cue_event_ids,
                tmin=-0.2,
                tmax=0.8,
                baseline=(None, 0),
                preload=True
            )

            # Save each subjectâ€™s epochs as .fif
            output_name = filename.replace('.raw', '-epo.fif')
            output_path = os.path.join(output_folder, output_name)
            epochs.save(output_path, overwrite=True)

            print(f"Saved epochs: {output_name}")
        except Exception as e:
            print(f"Error processing {filename}: {e}")
```

### ALIGNING ALL SAMPLES  TO HAVE UNIFIED CHANNEL NUMBER ( From N-> 149 -> 128)


```python
import mne
import os

input_folder = r"C:\Modma_dataset\EEG_128channels_ERP_lanzhou_2015\EEG_128channels_ERP_lanzhou_2015\Epochs_output"
output_folder = os.path.join(input_folder, "Epochs_aligned")
os.makedirs(output_folder, exist_ok=True)

# Use previously found values
reference_file = "02010006erp 20150528 1007-epo.fif"
ref_channel_order = ref_epochs.ch_names

for fname in os.listdir(input_folder):
    if fname.endswith("-epo.fif") and fname != reference_file:
        fpath = os.path.join(input_folder, fname)
        try:
            epochs = mne.read_epochs(fpath, preload=True, verbose=False)
            current_chs = set(epochs.ch_names)
            ref_chs = set(ref_channel_order)

            # Drop extra channels not in reference
            extra_chs = current_chs - ref_chs
            if extra_chs:
                print(f"âž– Dropping {len(extra_chs)} extra channels in {fname}")
                epochs.drop_channels(list(extra_chs))

            # Reorder to match reference
            intersecting_chs = [ch for ch in ref_channel_order if ch in epochs.ch_names]
            epochs.reorder_channels(intersecting_chs)

            # Save aligned
            out_path = os.path.join(output_folder, fname)
            epochs.save(out_path, overwrite=True)
            print(f"âœ… Saved aligned: {fname}")

        except Exception as e:
            print(f"âŒ Failed processing {fname}: {e}")
```

#### Checking Shape ( X,Y,Z)  of each files in the Aligned folder 

```python
import mne
from pathlib import Path

# Set the aligned folder path
aligned_dir = Path(r"C:\Modma_dataset\EEG_128channels_ERP_lanzhou_2015\EEG_128channels_ERP_lanzhou_2015\Epochs_output\Epochs_aligned")

# Reference file for checks
ref_path = aligned_dir / "02010006erp 20150528 1007-epo.fif"
ref_epochs = mne.read_epochs(ref_path, preload=False)
ref_ch_names = ref_epochs.ch_names
ref_shape = ref_epochs.get_data().shape

print(f"\nðŸ” Reference file: {ref_path.name}")
print(f"ðŸ“ Expected shape: {ref_shape}")
print(f"ðŸ§  Expected channels: {len(ref_ch_names)} channels\n")

# Loop through all aligned .fif files
for fpath in sorted(aligned_dir.glob("*-epo.fif")):
    try:
        epochs = mne.read_epochs(fpath, preload=False, verbose=False)

        # Check shape
        shape_match = epochs.get_data().shape == ref_shape

        # Check number of channels
        ch_count_match = len(epochs.ch_names) == len(ref_ch_names)

        # Check channel names and order
        ch_names_match = epochs.ch_names == ref_ch_names

        if shape_match and ch_count_match and ch_names_match:
            print(f"âœ… {fpath.name} passed all checks.")
        else:
            print(f"âš ï¸ {fpath.name} failed:")
            if not shape_match:
                print(f"   - Shape mismatch: {epochs.get_data().shape}")
            if not ch_count_match:
                print(f"   - Channel count mismatch: {len(epochs.ch_names)}")
            if not ch_names_match:
                print(f"   - Channel names/order mismatch")

    except Exception as e:
        print(f"âŒ Error loading {fpath.name}: {e}")
```

#### Since E1-E128 are not in sorted order , sorting channnel number in place 

```python
import os
import re
import mne
from natsort import natsorted

# Path to your folder
folder_path =r"C:\Modma_dataset\EEG_128channels_ERP_lanzhou_2015\EEG_128channels_ERP_lanzhou_2015\filtered128_folder"

# Natural sort helper
def natural_sort(ch_names):
    return natsorted(ch_names, key=lambda x: int(re.search(r'\d+', x).group()))

# Iterate over all FIF files and sort channels
for file in os.listdir(folder_path):
    if file.endswith(".fif"):
        file_path = os.path.join(folder_path, file)
        raw = mne.read_epochs(file_path, preload=True)

        # Natural sort channel names
        sorted_ch_names = natural_sort(raw.ch_names)

        # Reorder channels
        raw.reorder_channels(sorted_ch_names)

        # Optional: save back
        raw.save(file_path, overwrite=True)

        print(f"{file}: {raw.ch_names}")  # Print to verify sorted
```

```python
import os
import pandas as pd
import mne

# Path to EEG and metadata
eeg_folder = r"C:\Modma_dataset\EEG_128channels_ERP_lanzhou_2015\EEG_128channels_ERP_lanzhou_2015\filtered128_folder"
metadata_csv = r"C:\Modma_dataset\EEG_128channels_ERP_lanzhou_2015\EEG_128channels_ERP_lanzhou_2015\Pre_available_files\subjects_information_EEG_128channels_ERP_lanzhou_2015.xlsx"  # <- Save as CSV

# Load metadata
meta_df = pd.read_excel(metadata_csv, dtype={"Subject_Id": str})

# Ensure Subject_Id has consistent format (leading zeros)
meta_df['Subject_Id'] = meta_df['Subject_Id'].str.zfill(8)

# Map filenames to metadata
for fname in os.listdir(eeg_folder):
    if fname.endswith("-epo.fif"):
        subject_id = fname[:8]  # Extract from filename
        fpath = os.path.join(eeg_folder, fname)

        try:
            epochs = mne.read_epochs(fpath, preload=False, verbose=False)
            subj_meta = meta_df[meta_df['Subject_Id'] == subject_id]

            if subj_meta.empty:
                print(f"âš ï¸ No metadata for {subject_id}")
            else:
                print(f"âœ… {fname} matched â†’ {subj_meta.iloc[0].to_dict()}")
                
        except Exception as e:
            print(f"âŒ Failed to read {fname}: {e}")
```
